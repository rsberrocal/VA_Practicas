b) Visualize the histogram of the image. What is the histogram representing for an image? 
The plotted histogram must accomplish these requirements (see image bellow as an example of histogram extraction):

    - Obtain the histogram with 8 and 32 bins. 

    - Visualize the histogram of each channel and the gray level image separately and compare them. Comment their differences.

Note: What type should be the image in order to compute the histogram? check using different bins for the histogram. 

Hint : the `exposure` module within the `skimage` library makes the histogram extraction easy!

Help: [matplotlib image tutorial](https://matplotlib.org/users/image_tutorial.html)

<img src='images_for_notebook/ex21a.png'>

```python
def show_histogram(image, bins):
    # Get every channel
    gray_image = rgb2gray(image)
    red_image = image[:,:,0]
    green_image = image[:,:,2]
    blue_image = image[:,:,1]    
    
    # Pass to float
    gray_image = img_as_float(gray_image)
    red_image = img_as_float(red_image)    
    green_image = img_as_float(green_image)
    blue_image = img_as_float(blue_image)    
    
    # For each channel take his histogram
    histo_gray = exposure.histogram(gray_image, nbins=bins)
    histo_r = exposure.histogram(red_image, nbins=bins)
    histo_g = exposure.histogram(green_image, nbins=bins)
    histo_b = exposure.histogram(blue_image, nbins=bins)

    # Creating the plot
    fig, ax = plt.subplots(ncols = 1, figsize = (20, 5))
    # Fix a width for all the bars
    width = 0.2
    
    bins_gray = np.arange(0,bins) 
    bins_r = bins_gray  + width
    bins_g = bins_gray  + width * 2 
    bins_b = bins_gray  + width * 3 

    # Printing each histogram
    ax.bar(bins_gray, histo_gray[0], width=0.2, color='y')
    ax.bar(bins_r, histo_r[0], width=0.2, color='r')
    ax.bar(bins_g, histo_g[0], width=0.2, color='g')
    ax.bar(bins_b, histo_b[0], width=0.2, color='b')
    
    # Set the legend
    legend = ['Gray','Red', 'Green', 'Blue']
    
    # Set the title of the histogram.
    ax.set_title('Histogram with ' + str(bins) + ' nbins')
    
    # Set the legend of the histogram.
    ax.legend(legend)
    
    fig.tight_layout()
    
    plt.show()
```

```python
'''
Histrogram show us all the frecuency for each different data
'''

show_histogram(original_face, 8)

show_histogram(original_face, 32)

'''
A simple vista lo que se ve es que tenemos más valores bajos (colores oscuros) que del resto.

Los pixeles más oscuros predominan los canales verde y azul mientras que en los más claros el que predomina es el rojo.


'''
```

### 2.4 Image processing with weighting and non-linear filters

a) Generate the following kernels and apply them to the `face.png`:

1. Gaussian kernel with sigma = 1
2. Gaussian kernel with a different sigma value. Which is the most adequate value for keeping the main objects and structures of the image?
3. Median filter.

Comment the effect of using different kinds of filters on the original image.

Hint: search for the skimage.filters function for creating the different filters.

```python
# Gaussian filter from sckit image lib with sigma 1
face_gaussian = np.copy(original_face)
face_gaussian = gaussian_each(face_gaussian, 1)


# Gaussian filter from sckit image lib with sigma 6
# Responder la pregunta
'''
A low sigma is ideal for maintaining structures.
The resulting image will not be so blurry and you will be able to see the objects easily.
'''
face_gaussian_diff = np.copy(original_face)
face_gaussian_diff = gaussian_each(face_gaussian_diff, 6)

# Creating noisy image from original image
noisy = random_noise(original_face, mode='s&p')
median = np.copy(noisy)
filters.median(noisy, out=median)

# Struct of plots
fig = plt.figure(figsize=(15,15))

ax = fig.add_subplot(4 ,4, 1)
ax.title.set_text('Gaussian with sigma 1')
plt.axis('off')
plt.imshow(face_gaussian, cmap='gray')

ax = fig.add_subplot(4 ,4, 2)
ax.title.set_text('Gaussian with sigma 6')
plt.axis('off')
plt.imshow(face_gaussian_diff)

ax = fig.add_subplot(4, 4, 3)
ax.title.set_text('Noisy Image')
plt.axis('off')
plt.imshow(noisy)

ax = fig.add_subplot(4, 4, 4)
ax.title.set_text('Median Image')
plt.axis('off')
plt.imshow(median)

plt.show()
'''
We can see that what stands out most is that the Gaussian filter makes the image blurry,
on the other hand, the medium filter does not manage to erase all the added noise.
'''
```

```python
# Function that will show n images with their titles
# Useful to print differents results
def show_images_edges(images, titles):
    
    # Struct of plots
    fig = plt.figure(figsize=(20,20))
    
    for i in range(len(images)):        
        ax = fig.add_subplot(1 ,len(images), i+1)
        ax.title.set_text(titles[i])
        plt.axis('off')
        plt.imshow(images[i], cmap='gray')
    
    plt.show()
```

```python
# Getting the image
original_logo = io.imread('images/logo.png')

# Need to convert to gray, because these tecniques uses a 2-D Image
logo = rgb2gray(original_logo)

# Calling each flter and getting their results
logo_roberts = filters.roberts(logo)
logo_prewitt = filters.prewitt(logo)
logo_sobel = filters.sobel(logo)
# Canny needs a sigma because it uses a Gaussian filter internally
logo_canny = feature.canny(logo, sigma=.5)

# Put all the results in an array and call a function to display it
titles = ['Original','Roberts', 'Prewitt', 'Sobel', 'Canny']
logos = [original_logo,logo_roberts, logo_prewitt,logo_sobel,logo_canny]

'''
a) The best technique is the Canny, it uses a Gaussian filter to smooth the image and reduce noise.
b) A smaller sigma is ideal, a larger one makes the details of the contour disappear.
c) It is not completely necessary, Only depends on the final result you are looking for.
   If it is necessary to represent and visualize it is necessary to normalize it,
   if it is an intermediate process it would not be necessary.
'''


show_images_edges(logos, titles)
```

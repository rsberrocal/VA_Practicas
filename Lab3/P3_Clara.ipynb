```python
from skimage.exposure import histogram
from skimage import img_as_float, io

# We define a function that takes an image,
# converts it into float and returns the
# list of its three histograms 
# (one for each channel)

def data(image, nbins):
    image = img_as_float(image)
    return [histogram(image[:,:,rgb],nbins)[0] for rgb in range(3)]
```

```python
scenes = [0]

for i in range(2, len(ic)):
    
    difference = [abs(data(ic[i], 8)[rgb] - data(ic[i-1], 8)[rgb]) for rgb in range (3)]
    
    if int(np.mean(difference)) > 10000:
        scenes.append(i-1)
        scenes.append(i)
        
scenes.append(i)

plots = len(scenes)
fig = plt.figure(figsize=(250,250))

for p in range(plots):
    
    ax = fig.add_subplot(plots, 2, p + 1)
    plt.imshow(ic[scenes[p]])
    
    if p % 2 == 0:
        ax.title.set_text("Begin of scene")
    else:
        ax.title.set_text("End of scene")
```
